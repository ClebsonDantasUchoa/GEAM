## Criação de uma função para pré-processamento dos dados

```python
def preprocess(data):
    new_df = pd.DataFrame()
    new_df['NU_INSCRICAO'] = data['NU_INSCRICAO']
    new_df['NOTA_FINAL'] = (data['NU_NOTA_CN']* 2 + data['NU_NOTA_CH'] + data['NU_NOTA_LC']* 1.5 + data['NU_NOTA_MT'] * 3 + data['NU_NOTA_REDACAO'] * 3)/10.5
    new_df = new_df.dropna(thresh=2)
    return new_df
```

## Criação da variável que recebe o retorno da função

```python
notas = preprocess(df)
notas.isnull().sum()
```

```python
notas.head()
```

## Criação e ajustes finais do arquivo de resposta

```python
notas = notas.sort_values('NOTA_FINAL',ascending=False)
notas.head(20).to_csv("answer.csv")
```
